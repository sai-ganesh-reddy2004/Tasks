```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Load the uploaded image
image_path = "/content/36b1f17f60f61e76766ef0c642fc8064e3b41986.jpg" #change path if different
image = cv2.imread(image_path)

```

```python
# Convert BGR (OpenCV format) to RGB (Matplotlib format)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert to Grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Apply Gaussian Blur
blurred = cv2.GaussianBlur(image, (151, 111), 0)

# Apply Edge Detection (Canny)
edges = cv2.Canny(image, 100, 200)
```

```python
# Display Images
fig, ax = plt.subplots(1, 4, figsize=(15, 5))
ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[1].imshow(gray, cmap="gray")
ax[1].set_title("Grayscale")
ax[2].imshow(blurred)
ax[2].set_title("Blurred")
ax[3].imshow(edges, cmap="gray")
ax[3].set_title("Edges (Canny)")
plt.show()
```
